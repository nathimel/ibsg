```python
import numpy as np
import pandas as pd
import plotnine as pn
```

```python
(
    pn.ggplot(
        pd.DataFrame(
            np.random.normal(loc=0, scale=1, size=int(1e4)),
            columns=["X"]
        ),
        pn.aes(
            x="X",
        ),
    )
    + pn.geom_density(fill="orange")
    + pn.theme_classic()
    + pn.ylim(0, 0.6)
)
```
